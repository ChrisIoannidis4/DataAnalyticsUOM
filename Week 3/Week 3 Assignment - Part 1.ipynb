## Part 1 - NBA Player Comparision

```python
# As usual, we begin by importing the packages we will need

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The data consists of the shot log for the NBA season 2016/17

shot = pd.read_csv("Assignment Data/Week 3/NBA shotlog_16_17.csv")
pd.set_option('display.max_columns', 100)
print(shot.columns.tolist())
shot.describe()
```

```python
x = shot['location_x']
y = shot['location_y']
shot['halfcourt_x'] =np.where(shot['location_x'] < 933/2, 933 - shot['location_x'],shot['location_x'])
shot['halfcourt_y'] =np.where(shot['location_x'] < 933/2, 500 - shot['location_y'],shot['location_y'])
```

```python
#define subsets:
KevinDurant = shot[shot['shoot_player']=='Kevin Durant']
DwightHoward = shot[shot['shoot_player']=='Dwight Howard']
DeAndreJordan = shot[shot['shoot_player']=='DeAndre Jordan']
RussellWestbrook = shot[shot['shoot_player']=='Russell Westbrook']
#print("KD: " + str(KevinDurant.shape)+ "\nDJ: " + str(DeAndreJordan.shape)+ "\nDH: " + str(DwightHoward.shape) + "\nRW: " + str(RussellWestbrook.shape))
```

## Westbrook - Durant Comparision

```python
#Russell Westbrook:
hxW = RussellWestbrook['halfcourt_x']
hyW = RussellWestbrook['halfcourt_y']
f = plt.figure(figsize=(94/6,50/6))
ax = f.add_subplot(121)
colors = np.where(RussellWestbrook['current_shot_outcome']=='SCORED','r',np.where(RussellWestbrook['current_shot_outcome']=='MISSED','b','g'))
ax = plt.scatter(hxW,hyW, s=10, c= colors, marker= '.')
plt.grid(True)
plt.xlim(500,950)
plt.ylim(0,500)
plt.title("Russell Westbrook", fontsize = 15)

#Kevin Durant:
hxD = KevinDurant['halfcourt_x']
hyD = KevinDurant['halfcourt_y']
ax = f.add_subplot(122)
colors = np.where(KevinDurant['current_shot_outcome']=='SCORED','r',np.where(KevinDurant['current_shot_outcome']=='MISSED','b','g'))
ax = plt.scatter(hxD,hyD, s=10, c= colors, marker= '.')
plt.grid(True)
plt.xlim(500,950)
plt.ylim(0,500)
plt.title("Kevin Durant", fontsize = 15)
```
